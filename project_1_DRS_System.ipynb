### Functools — Higher-order functions and operations on callable objects in Python

The functools module provides important utilities for such higher order functions.

### threading
-  a thread is defined as the smallest unit of execution with the independent set of instructions
During threading, different processors run on a single program and each one of them performs
an independent task simultaneously. threading is used to create threds that can work together
- bascically it can be used to make one part of code to sleep for some time so that, 
another part is running and after sleep time is over then gets back and runs remaining part.
- A daemon thread will shut down immediately when the program terminates.
- Without daemon threads, you'd have to keep track of them, and tell them to exit, before your program can completely quit. By     setting them as daemon threads, you can let them run and forget about them, and when your program quits, any daemon threads     are killed automatically.
- https://www.knowledgehut.com/blog/programming/threading-in-python
